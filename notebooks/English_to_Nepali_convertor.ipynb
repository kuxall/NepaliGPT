```python
import pandas as pd
import requests
import time
from tqdm import tqdm
from time import perf_counter
start = perf_counter()
def translate(src, trg, word):
    url = f"https://translate.googleapis.com/translate_a/single?client=gtx&sl={src}&tl={trg}&dt=t&dt=bd&dj=1&q={word}"
    response = requests.get(url)
    if response.ok:
        result = response.json()
        return result["sentences"][0]["trans"]
    else:
        print("Translation request failed with status code:", response.status_code)
        return None

# Read the CSV file

df = pd.read_csv(filename, encoding="utf-8")
df
```

```python

df_cleaned = df[df["title"].notnull()].reset_index(drop=True)

translate_list = []

# Translate each entry
for number in tqdm(range(len(df_cleaned))):
    title = df_cleaned["title"][number]
    # print(title)
    content = df_cleaned["content"][number]
    # print(content)
    url = df_cleaned["URL"][number]
    # print(url)
    try:
        translated_title = translate(src="en", trg="ne", word=title)
        translated_content = translate(src="en", trg="ne", word=content)
        translate_list.append({"title": translated_title, "content": translated_content})
    except Exception as e:
        print(f"Translation failed for entry {number}: {e}")
        # translate_list.append({"title": None, "content": None, "url": url})

    # Sleep for a short interval to avoid API rate limiting
    time.sleep(0.1)
print(type(translate_list))

# Create a DataFrame and save the results to a CSV file
df2 = pd.DataFrame(translate_list)
print(type)
print(f"Translation and CSV writing took {perf_counter() - start:.2f} seconds")
```
